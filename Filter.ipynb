```python
#Q5.Write a Pandas program to remove the duplicates from 'WHO region' column of World alcohol consumption dataset

df.drop_duplicates(subset='WHO region') # to make it permanent we can include inplace = True
df.shape
```

```python
#Q6.Write a Pandas program to find out the alcohol consumption of a given year from the world alcohol consumption dataset

df.groupby('Year')['Display Value'].sum().reset_index()
```

```python
#Q7.Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset

df.loc[df['Year']==1987]
```

```python
#Q8.Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset

df.loc[(df['Year']==1985) & (df['WHO region']=='Americas')]
```

```python
#Q9.Write Pandas program to find out the alcohol consumption details in the year '1986' 
#where WHO region is 'Western Pacific' and country is 'VietNam' from the world alcohol consumption dataset

df.loc[(df['Year']==1986) & (df['WHO region']=='Western Pacific') & (df['Country']=='Viet Nam') ]
```

```python
#Q10.Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' 
#where WHO region is 'Americas' from the world alcohol consumption dataset

df.loc[(df['Year']==1986) & (df['WHO region']=='Americas')]
```

```python
#Q12.Write a Pandas program to find out the 'WHO region, 'Country', 'Beverage Types' in the year '1986' or '1989' 
#where WHO region is 'Americas' or 'Europe' from the world alcohol consumption dataset.

df.loc[((df['Year']==1986) | (df['Year']==1989)) &  ((df['WHO region']=='Americas')  | (df['WHO region']=='Europe'))][['WHO region', 'Country', 'Beverage Types','Year']]
```

```python
#Q13.Write a Pandas program to find out the records where consumption of beverages per person average >=5 
#and Beverage Types is Beer from world alcohol consumption dataset

df.loc[(df['Display Value']>=5)  & (df['Beverage Types']=='Beer')]
```

```python
#Q15.Write a Pandas program to filter the specified columns and records by range from world alcohol consumption dataset

df.loc[0:4, ["WHO region", "Beverage Types"]]
```

```python
#Q16.Write a Pandas program to filter those records where WHO region contains "Ea" substring 

df.loc[df['WHO region'].str.contains('Ea')]
```

```python
#Q17.Write a Pandas program to filter those records where WHO region matches with multiple values 
#(Africa, Eastern Mediterranean, Europe) from world alcohol consumption dataset

df.loc[df["WHO region"].isin(["Africa", "Eastern Mediterranean", "Europe"])].head(10)
```

```python
#Q18.Write a Pandas program to filter those records which not appears in a given list from world alcohol consumption dataset.

regions=['Americas','Europe']

df.loc[~df['WHO region'].isin(regions)].head(10)
```

```python
#Q19.Write a Pandas program to filter all records where the average consumption of beverages per person from .5 to 2.50 

df.loc[(df['Display Value'] > 0.5) & (df['Display Value'] <= 2.50)]
```

```python
df.loc[df['Display Value'].between(0.5,2.50)]
```

```python
#Q21.Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 

df.filter(regex='0$', axis=0)
```
