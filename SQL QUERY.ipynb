```python
#Q6.Write Pandas program to display first,last name,salary and department number for employees whose first name starts with 'S'

employee[employee['first_name'].str.startswith('S')][['first_name','last_name','salary','department_id']]
```

```python
#Q7.Write a Pandas program to display the first, last name, salary and department number for those employees whose first name does not contain the letter 'M'

employee[~employee['first_name'].str.contains('M')][['first_name','last_name','salary','department_id']]
```

```python
#Q8. Write a Pandas program to display the first name, last name, salary and department number in ascending order by department number.

employee.sort_values(by='department_id',ascending=True)[['department_id','first_name','last_name','salary']]
```

```python
#Q10. Write a Pandas program to display the first name, last name, salary and manger id where manager ids are null.

employee[employee['manager_id'].isna()][['first_name','last_name','salary','manager_id']]
```

```python
#Q11. Write a Pandas program to display the first name, last name, salary and manger id where manager ids are not null.

employee[~employee['manager_id'].isna()][['first_name','last_name','salary','manager_id']]
```

```python
#Q21. Write a Pandas program to display the ID for those employees who did two or more jobs in the past.

result=job_history.groupby('employee_id').size().reset_index(name='No of Jobs')

result[result['No of Jobs']>1]['employee_id'].reset_index()
```
