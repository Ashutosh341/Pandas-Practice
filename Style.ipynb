```python
import pandas as pd
import numpy as np
```

```python
#Q1.Create a dataframe of ten rows, four columns with random values. 
#Write a Pandas program to highlight the negative numbers red and positive numbers black. 

df= pd.DataFrame(np.random.randn(10,4),columns=list('WXYZ'))
df
```

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color
print("\nNegative numbers red and positive numbers black:")
df.style.applymap(color_negative_red)
```

```python
df.style.applymap(lambda x:'color:red' if x<0 else 'color:black')
```

```python
#Q2.Create a dataframe of ten rows, four columns with random values. 
#Convert some values to nan values. Write a Pandas program which will highlight the nan values.

df.iloc[0,2]=np.nan # row 0 2nd column
df.iloc[5,0]=np.nan
df.iloc[6,1]=np.nan
df.iloc[7,3]=np.nan
df.iloc[8,3]=np.nan
df.iloc[4,3]=np.nan


df
```

```python
df.style.highlight_null(null_color='red')
```

```python
#Q3.Create dataframe of ten rows, four columns with random values.
#Write a Pandas program to highlight the maximum value in each column

df.style.highlight_max(color = 'lightgreen', axis = 0) #axis = 1 for each row and axis=0 for each column
```

```python
#Q4.Create dataframe of ten rows, four columns with random values.
#Write a Pandas program to highlight the min value in each column

df.style.highlight_min(color = 'yellow', axis = 0) #axis = 1 for each row and axis=0 for each column
```

```python
#Q5. Create a dataframe of ten rows, four columns with random values.
#Write a Pandas program to highlight the maximum value in last two columns.

df.style.highlight_max(color = 'yellow', axis = 0,subset=['Y','Z'])
```

```python
#Q7. Create a dataframe of ten rows, four columns with random values.
#Write a Pandas program to highlight dataframe's specific columns

df.style.applymap(lambda x:'background-color:yellow',subset=['X','Y'])
```

```python
#Q10. Create a dataframe of ten rows, four columns with random values. 
#Write a Pandas program to highlight the entire row in Yellow where a specific column value is greater than 0.5.

def highlight(s):
    
    if s.X > 1:
        return ['background-color: yellow'] * len(s)
    else:
        return ['background-color: white'] * len(s)
df.style.apply(highlight, axis=1)
```
