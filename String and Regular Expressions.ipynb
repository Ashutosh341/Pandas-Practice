```python
#Q24.Write a Pandas program to extract email from a specified column of string type of a given DataFrame

df = pd.DataFrame({
    'name_email': ['Alberto Franco af@gmail.com','Gino Mcneill gm@yahoo.com','Ryan Parkes rp@abc.io', 'Eesha Hinton', 'Gino Mcneill gm@github.com']
    })

df
```

```python
import re

def findingEmail(text):
    result = re.findall(r"[A-Za-z0-9]*@\w*\.\w*",text)
    
    return ','.join(result)

df['email'] =df['name_email'].apply(lambda x:findingEmail(x))
df
```

```python
#Q25.Write a Pandas program to extract hash attached word from twitter text from the specified column of a given DataFrame

df = pd.DataFrame({
    'tweets': ['#Obama says goodbye','Retweets for #cash','A political endorsement in #Indonesia', '1 dog = many #retweets', 'Just a simple #egg']
    })

import re

def findingHashtags(text):
    result = re.findall(r"#\w*",text)
    
    return ','.join(result)

df['hastags'] =df['tweets'].apply(lambda x:findingHashtags(x).split('#')[1])
df
```

```python
#Q26. Write a Pandas program to extract word mention someone in tweets using @ from the specified column of a given DataFrame

df = pd.DataFrame({
    'tweets': ['@Obama says goodbye','Retweets for @cash','A political endorsement in @Indonesia', '1 dog = many #retweets', 'Just a simple #egg']
    })

import re

def findingmentions(text):
    result = re.findall(r"@\w*",text)
    
    return ' '.join(result)

df['Mentions'] =df['tweets'].apply(lambda x:findingmentions(x))
df
```

```python
#Q27. Write a Pandas program to extract only number from the specified column of a given DataFrame.

df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'address': ['7277 Surrey Ave.','920 N. Bishop Ave.','9910 Golden Star St.', '25 Dunbar St.', '17 West Livingston Court']
    })

def findingnumbers(text):
    result = re.findall(r"[0-9]+",text)
    
    return ' '.join(result)

df['numbers_in_address'] =df['address'].apply(lambda x:findingnumbers(x))
df
```

```python
#Q37. Write a Pandas program to extract only words from a given column of a given DataFrame.

df = pd.DataFrame({
    'company_code': ['Abcd','EFGF', 'zefsalf', 'sdfslew', 'zekfsdf'],
    'date_of_sale': ['12/05/2002','16/02/1999','05/09/1998','12/02/2022','15/09/1997'],
    'address': ['9910 Surrey Ave.','92 N. Bishop Ave.','9910 Golden Star Ave.', '102 Dunbar St.', '17 West Livingston Court']
})

def findingwords(text):
    result = re.findall(r"[A-Za-z]*",text)
    
    return ' '.join(result)

df['onlywords'] =df['address'].apply(lambda x:findingwords(x))
df
```

```python
#Q41. Write a Pandas program to remove the html tags within the specified column of a given DataFrame
df = pd.DataFrame({
    'company_code': ['Abcd','EFGF', 'zefsalf', 'sdfslew', 'zekfsdf'],
    'date_of_sale': ['12/05/2002','16/02/1999','05/09/1998','12/02/2022','15/09/1997'],
    'address': ['9910 Surrey <b>Avenue</b>','92 N. Bishop Avenue','9910 <br>Golden Star Avenue', '102 Dunbar <i></i>St.', '17 West Livingston Court']
})

def remove_tags(text):
    result = re.sub(r"<\w?\W?\w?>",'',text)
    
    return ''.join(result)

df['Address_without_tags'] =df['address'].apply(lambda x:remove_tags(x))
df
```
